```python
df_january['trip_duration'] = df_january['tpep_dropoff_datetime'] - df_january['tpep_pickup_datetime']

df_filtered = df_january[(df_january['trip_duration'] >= pd.Timedelta(minutes=1)) & (df_january['trip_duration'] <= pd.Timedelta(minutes=60))]
```

```python
from sklearn.feature_extraction import DictVectorizer

# Convertir les colonnes en chaînes de caractères
df_january['PULocationID'] = df_january['PULocationID'].astype(str)
df_january['DOLocationID'] = df_january['DOLocationID'].astype(str)
```

```python
# Transformer le DataFrame en une liste de dictionnaires
data = df_january[['PULocationID', 'DOLocationID']].to_dict('records')
```

```python
# Initialiser le DictVectorizer avec sparse=True (par défaut)
vectorizer = DictVectorizer(sparse=True)

# Ajuster le DictVectorizer et obtenir la matrice de caractéristiques creuse
feature_matrix = vectorizer.fit_transform(data)
```

```python
# Afficher la matrice de caractéristiques
print(feature_matrix)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
# Ajuster le DictVectorizer et obtenir la matrice de caractéristiques creuse
feature_matrix = vectorizer.fit_transform(data)

# Préparer la variable de réponse (durée)
# Assurez-vous que la colonne 'duration' existe dans votre DataFrame
duration = df_january['trip_duration'].values

# Initialiser le modèle de régression linéaire
model = LinearRegression()

# Entraîner le modèle
model.fit(feature_matrix, duration)
```

```python
# Prédire les valeurs de durée sur les données d'entraînement
predicted_duration = model.predict(feature_matrix)

# Calculer l'erreur quadratique moyenne (RMSE)
rmse = np.sqrt(mean_squared_error(duration, predicted_duration))

# Afficher le RMSE
print(f"RMSE: {rmse}")
```
